```python
def freq(word:str) -> dict:
  word = word.lower()
  space_cnt = 0
  freq = {}
  for w in word:
    if w == ' ':
      if space_cnt == 0: # cnt가 0이면 전의 단어 까지 포함
        space_cnt += 1
      space_cnt+=1
    if w in freq:
      freq[w] += 1
    else:
      freq[w] = 1
  freq['단어수'] = space_cnt
  return freq

word = input("영문장 입력 : ")

print(freq(word))
```

```python
def freq(word:list) -> dict:
  space_cnt = len(word)
  freq = {}
  for w in word:
    w = w.lower()
    for i in w:
      if i in freq:
        freq[i] += 1
      else:
        freq[i] = 1
  freq['단어수'] = space_cnt
  return freq

word = input("영문장 입력 : ").split()

print(freq(word))
```

```python
def search(lst:list, first_name:str) -> int: # 리스트와 성을 보내주면 해당 성의 카운트를 반환
  cnt = 0
  for name in lst:
    if name[0] == first_name:
      cnt += 1
  return cnt

def search_loop(lst:list, name:str) -> int:
  cnt = 0
  for name_item in lst:
    if name_item == name:
      cnt += 1
  return cnt

def duplication_sort(lst:list) -> list:
  lst.sort()
  for i in lst:
    for j in lst:
      if i == j:
        del lst[lst.index(j)]
  return lst


name_lst = input("이름을 입력 : ").split(",")

search_first_name = input("성(first_name)을 입력 : ")
print(search(name_lst, search_first_name))
search_first_name = input("성(first_name)을 입력 : ")
print(search(name_lst, search_first_name))

search_name = input("이름이 몇번 반복되는지? : ")
print(search_loop(name_lst, search_name))

print("중복 삭제, 정렬 : ", duplication_sort(name_lst))
```

```python
def format_num(num:str) -> str:
  num = str(num)
  for i in range(len(num), 0, -3):
      num = num[:i] + ',' + num[i:]
  if num[-1] == ',':
    num = num[:-1]
  return num

n = input("숫자 입력 : ")
print("금액 출력 : ",format_num(n))

```

```python
def BinToDecimal(Bin:int) -> int:
  res = 0
  cnt = len(str(Bin))
  for i in range(len(str(Bin))):
    cnt -= 1
    if str(Bin)[i] == '1':
      res += (2 ** (cnt))

  return res

def BinToHex(Bin:int) -> str:
  res = ''
  dec = BinToDecimal(Bin)
  while(dec > 0):
    tmp = dec % 16
    if tmp < 10:
      res = str(tmp) + res
    else:
      res = chr(55 + tmp) + res
    dec = dec // 16
  return res

def HexToBin(Hex:str) -> str:
  res = ''

  for i in Hex:
    if not(49 <= ord(i) <= 57): # true이면 숫자아님
      i = (10 + (ord(i) - 65))
    bin_val = ''
    for j in range(4):
      bin_val = str(int(i) % 2) + bin_val
      i = int(i) // 2
    res += bin_val
  return res

print(HexToBin('A5'))

print(BinToDecimal(1110010))

print(BinToHex(1110010))
```

```python
def encoding(st:str) -> str:
  res = ''
  tmp = {}
  for i in st:
    if i in tmp:
      tmp[i] += 1
    else:
      tmp[i] = 1
  for k, v in tmp.items():
    res += k+str(v)
  return res



def Decoding(st:str) -> str:
  res = ''
  tmp = ''
  st.split()
  for i in st:
    if i.isalpha():
      tmp += i
    else:
      res += tmp * int(i)
      tmp = ''
  return res

s = input("입력 : ")
res = (encoding(s))
print(res)
print(Decoding(res))
```

```python
def sortt(s:str) -> str:
  num_lst = [int(i) for i in s if i.isdigit()]
  chr_lst = [str(j) for j in s if j.isalpha()]

  res = ''.join(sorted(chr_lst)) + str(sum(num_lst))
  return res

st = input("입력 : ")
print(sortt(st))
```
